```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
from glob import glob
from tqdm import tqdm
```

```python
#figure 2B 1
refname='figure_2b_nopba'
filename='./videos/'+refname+'/'+refname
# rand=np.random.randint(0,10000)
deltaT = 0.005
timeStep = 35000
totalTime = deltaT*timeStep




N=2
Lx=20
Ly=500
R = 1
m = 1
A = 0.01
B = 3
tau = 1


updateT=tau
updateF=round(updateT/deltaT)



Nr = 1/2
deltheta=np.pi/8
numagents=N
tau_fut=6
agentradius=R

numsamp=50





# number of agent in each group
g1 = int(round(N * Nr))
# g1=1
g2 = N - g1


filestart='./2agents_nofsm/*.pickle'
files=glob(filestart)

for file in [files[0]]:
    with open(file,'rb') as f:
        alldata=pickle.load(f)
    xData=np.array(alldata['xData'])
    yData=np.array(alldata['yData'])-250
    uData=np.array(alldata['uData'])
    vData=np.array(alldata['vData'])
    desdirs=np.array(alldata['desdirs'])

start=0
end=int(40/deltaT)
for i in tqdm(range(start,end,50)):
    plt.figure()
    for ag in range(N):
        xpos=(xData[i][ag])
        ypos=(yData[i][ag])
        vx=uData[i][ag]
        vy=vData[i][ag]
        desx=desdirs[i][ag,0]
        desy=desdirs[i][ag,1]

        plt.gca().add_patch(plt.Circle((xpos,ypos),1,fc='red' if ag<g1 else 'blue'))
        plt.arrow(xpos,ypos,vx,vy)
        plt.arrow(xpos,ypos,desx,desy,color='white',alpha=0.5)

        if xpos<1:
            plt.gca().add_patch(plt.Circle((Lx+xpos,ypos),1,fc='red' if ag<g1 else 'blue',alpha=1))
            plt.arrow(Lx+xpos,ypos,vx,vy,alpha=1)
        if xpos>Lx-1 or ypos>Ly-1:
            plt.gca().add_patch(plt.Circle((xpos-Lx,ypos),1,fc='red' if ag<g1 else 'blue',alpha=1))
            plt.arrow(xpos-Lx,ypos,vx,vy,alpha=1)


        

    plt.xlim(0,Lx)
    plt.ylim(-7,7)
    # plt.axhline(0)
    # plt.axhline(Ly)
    plt.title(refname+'; t='+format(i*deltaT,'.2f'))
    plt.gca().set_aspect(1)
    plt.savefig(filename+str(i)+'_'+'.png',facecolor='white')
    plt.close()
```
